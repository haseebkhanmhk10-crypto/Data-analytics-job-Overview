**Task 1:** Exploring and Visualizing a Simple Dataset

**Objective:**
 Understand how to read, summarize, and visualize a dataset using pandas, matplotlib, and seaborn..

**Dataset:**
 Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv('/content/iris.csv')

print("Dataset loaded successfully!")
```

```python
print("Shape of dataset:", iris_df.shape)
print("Column names:", iris_df.columns.tolist())

print("\nFirst 5 rows:")
print(iris_df.head())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df)

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)

plt.savefig("scatter_plot.png", dpi=300)
plt.show()
```

This scatter plot shows the relationship between sepal length and sepal width.
Different species are represented using different colors.

```python
plt.figure(figsize=(8,6))
sns.histplot(iris_df['sepal_length'], bins=10, kde=True)

plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.savefig("histogram.png", dpi=300)
plt.show()
```

The histogram shows how sepal length values are distributed across the dataset.

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='petal_length', data=iris_df)

plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')

plt.savefig("boxplot.png", dpi=300)
plt.show()
```

The box plot shows the spread and distribution of petal length for each species.
It helps identify median values and any outliers.
